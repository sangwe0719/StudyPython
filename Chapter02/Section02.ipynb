```python
value = "파인애플"
match value:
    case '망고' :
        result = 'mango'
    case '사과' :
        result = 'apple'
    case '바나나' :
        result = 'banana'
    case _:
        result = '주문하신 과일이 없습니다.'
print(result) 
```

```python
def sum_sub_mul(x,y,z):
    hap = x + y + z 
    cha = x - y - z
    mul = x * y * z
    return hap,cha,mul
sum_sub_mul(10,3,2)
```
